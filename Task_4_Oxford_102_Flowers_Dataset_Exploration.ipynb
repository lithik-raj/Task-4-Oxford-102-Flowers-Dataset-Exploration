```python
import scipy.io
import numpy as np

# Load flower labels
labels = scipy.io.loadmat("imagelabels.mat")["labels"][0]

print("Oxford-102 Flowers Dataset")
print("---------------------------")
print("Total Images :", len(labels))
print("Total Classes:", len(np.unique(labels)))
```

```python
from PIL import Image
import os

image_folder = "jpg"

widths = []
heights = []

image_files = os.listdir(image_folder)

for image_name in image_files:

    image_path = os.path.join(image_folder, image_name)

    img = Image.open(image_path)

    width, height = img.size

    widths.append(width)
    heights.append(height)

print("Image Resolution Statistics")
print("---------------------------")
print("Average Width :", round(np.mean(widths), 2))
print("Average Height:", round(np.mean(heights), 2))
print("Minimum Width :", min(widths))
print("Maximum Width :", max(widths))
```

```python
import random
import matplotlib.pyplot as plt

sample_indices = random.sample(range(len(image_files)), 5)

for idx in sample_indices:

    image_name = image_files[idx]

    image_path = os.path.join(image_folder, image_name)

    img = Image.open(image_path)

    flower_label = labels[idx]

    plt.figure(figsize=(5,5))
    plt.imshow(img)
    plt.title("Flower Class: " + str(flower_label))
    plt.axis("off")
    plt.show()
```

```python
from PIL import Image
import os
import numpy as np

image_folder = "jpg"
image_files = sorted(os.listdir(image_folder))

widths = []
heights = []

for image_file in image_files:

    img = Image.open(os.path.join(image_folder, image_file))

    width, height = img.size

    widths.append(width)
    heights.append(height)

print("Image Resolution Statistics")
print("---------------------------")
print("Average Width :", round(np.mean(widths), 2))
print("Average Height:", round(np.mean(heights), 2))
print("Minimum Width :", min(widths))
print("Maximum Width :", max(widths))
```

```python
import random
import matplotlib.pyplot as plt

sample_images = random.sample(image_files, 5)

for image_name in sample_images:

    index = image_files.index(image_name)

    flower_label = labels[index]

    img = Image.open(os.path.join(image_folder, image_name))

    plt.figure(figsize=(5,5))
    plt.imshow(img)
    plt.title("Flower Class: " + str(flower_label))
    plt.axis("off")
    plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.hist(labels, bins=102)

plt.title("Flower Class Distribution")
plt.xlabel("Class Number")
plt.ylabel("Number of Images")

plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.scatter(widths, heights, alpha=0.5)

plt.title("Image Resolution Distribution")
plt.xlabel("Width (pixels)")
plt.ylabel("Height (pixels)")

plt.show()
```
